```python
from common.database.adatabase import ADatabase
import pandas as pd
```

```python
market = ADatabase("market")
```

```python
market.connect()
cards = market.retrieve("pokemon")
market.disconnect()
```

```python
cards = cards.sort_values("ticker").groupby(["ticker","rarity","date"]).mean().reset_index().drop("directLow",axis=1).dropna()
```

```python
cards["spread"] = (cards["mid"] - cards["low"]) / cards["adjclose"]
```
